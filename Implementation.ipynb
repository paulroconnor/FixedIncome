## Bond

```python
b = Bond(facevalue = 100, couponrate = 0.20, frequency = 'Monthly', maturitydate = '2025-01-13', valuationdate = '2024-06-18', compounding = 'Semi-Annual', convention = 'Actual/Actual', region = 'United States')
```

```python
b.price
```

```python
b.valuationtable
```

```python
b.plot('Present Value')
```
